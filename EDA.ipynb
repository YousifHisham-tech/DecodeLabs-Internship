# Start cleaning steps
### Remove Unused columns 

```python
print("The number of unique orders is:", df['OrderID'].nunique()) # Check if the customer has multiple orders.
# This will be useful for customer segmentation and understanding customer behavior.
# But here we can see that each order has a unique OrderID, so we can use OrderID as a unique identifier for each transaction.
# This means that this feature isn't useful so we will drop it
df.drop('OrderID', axis=1, inplace=True)
```

```python
print("The number of unique customers is:", df['CustomerID'].nunique()) # Check how many unique customers we have in the dataset. 
# The number of customers is very high on the data 
# This means that we have a lot of customers with only one order, which makes it impossible to do customer segmentation and understanding customer behavior.
# We will drop this feature as well because it won't be useful for our analysis.
df.drop('CustomerID', axis=1, inplace=True)
```

```python
print("The number of unique tracking numbers is:", df['TrackingNumber'].nunique()) # Check how many unique tracking numbers we have in the dataset.
# This number is very high so it is not useful for our analysis and we will drop it as well.
df.drop('TrackingNumber', axis=1, inplace=True)
```

```python
print ("The number of unique coupon codes is:",df['CouponCode'].nunique(), end="\n\n") # Check how many unique coupon codes we have in the dataset.
print ("The distrbution of coupon codes is:") # Check the distribution of the coupon codes in the dataset.
pd.DataFrame(df['CouponCode'].value_counts()) # Check the distribution of the coupon codes in the dataset.
```

```python
print ("The number of shipping address is:", df['ShippingAddress'].nunique(), end="\n\n")
print ("The distribution of the shipping address is:")
pd.DataFrame(df['ShippingAddress'].value_counts(sort=True))
# I will let it may I extract some features
```
